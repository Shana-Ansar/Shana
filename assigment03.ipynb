```python
import pandas as pd
```

```python
df = pd.read_excel("iris.xls")
print(df.columns)
```

```python
df.mean(numeric_only=True)
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='Classification')
plt.show()
```

```python
df_melt = pd.melt(df, "Classification", var_name="measurement")
plt.figure(figsize=(10, 6))
sns.boxplot(x="measurement", y="value", hue="Classification", data=df_melt)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.violinplot(x="Classification", y="SL", data=df)
plt.title("Sepal Length")
plt.subplot(1, 2, 2)
sns.violinplot(x="Classification", y="SW", data=df)
plt.title("Sepal Width")
plt.tight_layout()
plt.show()
```
